## Data Cleaning:

For Duplicated Rows:

```python
#Identifying duplicate rows

duplicated_data = df[df.duplicated(keep=False)]
print("Number of duplicate rows:", df.duplicated().sum())
duplicated_data.sort_values(by="ResponseId", ascending=True, inplace=True)
duplicated_data.head()
```

```python
#Removing duplicated data

new_df = df.drop_duplicates()
print("Number of duplicate rows:", new_df.duplicated().sum())
```

For Missing Values:

```python
#Identifying missing values column wise

missing_data = new_df.isnull()
for column in missing_data.columns.values.tolist():
  print(column)
  print(missing_data[column].value_counts())
  print("")
```

For missing data, we will impute or drop data according to the situation and the missing data pattern.

```python
#Visualizing missing data as a heatmap

plt.figure(figsize=(20, 12))
ax=plt.axes()
sns.heatmap(missing_data, cbar=False, cmap='viridis', ax=ax)
plt.title('Missing Data Heatmap')
plt.xlabel('Columns')
plt.ylabel('Missing Values')
plt.show()
```

For Normalizing Datasets:

```python
#Applying Min-Max Scaling to ConvertedCompYearly

min_value = new_df['ConvertedCompYearly'].min()
max_value = new_df['ConvertedCompYearly'].max()
new_df['ConvertedCompYearly_MinMiax'] = (new_df['ConvertedCompYearly'] - min_value) / (max_value - min_value)
```

```python
#Applying Log Transform to ConvertedCompYearly to reduce skewness

new_df['ConvertedCompYearly_Log'] = np.log(new_df['ConvertedCompYearly'])
```

```python
#Creating Experience Level based on YearsCodePro

new_df['YearsCodePro'] = pd.to_numeric(new_df['YearsCodePro'], errors='coerce')
def assign_experience_level(years):
    if pd.isna(years):
      return 'Unknown'
    elif years <= 2:
        return 'Beginner'
    elif 3 <= years <= 5:
        return 'Intermediate'
    else:
        return 'Advanced'

new_df['ExperienceLevel'] = new_df['YearsCodePro'].apply(assign_experience_level)
new_df[['YearsCodePro', 'ExperienceLevel']].tail(15)
```
